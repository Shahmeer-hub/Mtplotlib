```python
#import matplotlab's pyplot as plt
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
#write your label for x-axis
plt.xlabel("Day")
```

```python
#write your label for y-axis
plt.ylabel("Temperature")
#write your title for the plot
plt.title("weather")
```

```python
plt.show(x,y)
```

```python
plt.plot(x,y, color='red', linewidth = 2)
```

```python
#to change the by default plotting we can pass aurguments to customize the plot.
plt.plot(x,y, color='red', linewidth=2, linestyle='dotted')
```

```python
plt.plot(x,y, color='green', linewidth=5, linestyle='dashed')
```

```python
plt.plot(y, 'o')
plt.show()
```

```python
#you can also customize the line
plt.plot(x,y, color='black', marker="o", linestyle='--')
```

```python
#if we put line style empty then it will show only markers
plt.plot(x,y, color='blue', marker="^", linestyle='', markersize=10)
```

```python
# LABEL PARAMETER

days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')
```

```python
# LEGEND

plt.plot(days, max_t, label='max')
plt.plot(days, min_t, label='min')
plt.plot(days, avg_t, label='average')

plt.legend(loc='best')
```

```python
# BAR CHART

import numpy as np
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

```python
# HORIZONTAL BARS
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

```python
#you can customize the width of the bar with width aurgument 
plt.bar(x, y, width = 0.1)
plt.show()
```
